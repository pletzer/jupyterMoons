# daily fluctuations of USD- NZD exchange rate

```python
import pandas
dataApr2016 = pandas.read_csv('DAT_MT_NZDUSD_M1_201604.csv', header=None)
```

```python
dates = dataApr2016[0]
times = dataApr2016[1]
nzd2usd = (dataApr2016[2] + dataApr2016[3] + dataApr2016[4] + dataApr2016[5])/4.
ts = []
for t in times:
    hour, minute = t.split(':')
    ts.append(float(hour) + float(minute)/60.)

```

```python
%matplotlib inline
from matplotlib import pyplot
pyplot.rcParams['figure.figsize'] = (15, 6)
pyplot.plot(ts, nzd2usd, 'o')
pyplot.ylabel('USD/NZD')
pyplot.xlabel('time of day')
```
