```python
import pandas
data = pandas.read_excel('https://www.reinz.co.nz/Media/Default/Statistic%20Documents/HPI/2016%20May/Summary%20of%20HPI%20for%20members.xlsx')
```

```python
print(data)
```

```python
%matplotlib inline
from matplotlib import pyplot

pyplot.figure(figsize=(12, 6))
colors = ['b', 'c', 'm', 'r', 'k',]
i = 0
for year in range(2000, 2017):
    months = data[data['Year'] == year]['Month']
    prices = data[data['Year'] == year]['Stratified Median']
    #priceIndex = data[data['Year'] == year]['HousePrice Index']
    c = colors[i % len(colors)]
    pyplot.plot(months, prices, c + '-')
    i += 1
```
